## Cleaning

```python
import re
import pandas as pd
import numpy as np

def clean_text(text):
    if pd.isnull(text) or str(text).strip().lower() == 'nan':
        return np.nan

    text = str(text).strip()

    text = re.sub(r'[^\w\s\-/]', '', text)
    text = text.replace("\n", " ").replace("\\n", " ")
    text = text.replace("\xa0", " ").replace("�", "")

    return text.strip()

for col in categorical_columns:
    df[col] = df[col].astype(str).apply(clean_text)

for col in categorical_columns:
    print(f'⭐ {col}: {df[col].nunique()} unique values')

```

```python
import pandas as pd
import re
from thefuzz import process, fuzz

# --- 1. تنظيف أولي ---
def clean_company_name(name):
    if pd.isna(name):
        return "Unknown"
    name = str(name).strip()
    name = re.sub(r'\s+', ' ', name)
    name = re.sub(r'[^\w\s&.-]', '', name)
    return name.title()

df['company_name'] = df['company_name'].apply(clean_company_name)

# --- 2. Fuzzy Matching  ---
unique_names = df['company_name'].unique()
mapping = {}

for name in unique_names:
    match = process.extractOne(name, mapping.keys(), scorer=fuzz.token_sort_ratio)
    if match and match[1] >= 80:
        mapping[name] = match[0]
    else:
        mapping[name] = name

df['company_name'] = df['company_name'].map(mapping)

# --- 3. show result---
print("number of unique company after cleaning : ", df['company_name'].nunique())
print(df['company_name'].value_counts().head(20))
```

```python
import pandas as pd
import re
from collections import Counter

company_stop_words = set([
    'inc', 'inc.', 'ltd', 'ltd.', 'llc', 'llp', 'co', 'co.', 'corp', 'corporation',
    'group', 'company', 'limited', 'plc', 'pvt', 'private', 'ag', 'sa', 'bv', 'gmbh',
    'kg', 'srl', 'spa', 'pty', 'nv', 'holdings', 'holding', 'international', 'global'
])

def tokenize_company(name):
    if pd.isnull(name):
        return []
    name = name.lower()
    name = re.sub(r'[^\w\s-]', '', name)
    name = name.replace('-', ' ')
    words = name.split()
    return [w for w in words if w not in company_stop_words]

all_company_words = []
for company in df['company_name'].unique():
    all_company_words.extend(tokenize_company(company))

company_word_counts = Counter(all_company_words)

filtered_company_keywords = [(w, c) for w, c in company_word_counts.items() if c > 2]
filtered_company_keywords = sorted(filtered_company_keywords, key=lambda x: x[1], reverse=True)[:300]

for word, count in filtered_company_keywords:
    print(f"{word}: {count}")

company_map = {
    'google': 'Google',
    'alphabet': 'Google',
    'microsoft': 'Microsoft',
    'apple': 'Apple',
    'ibm': 'IBM',
    'meta': 'Meta',
    'facebook': 'Meta',
    'amazon': 'Amazon',
    'aws': 'Amazon',
    'tesla': 'Tesla',
    'netflix': 'Netflix',
    'adobe': 'Adobe',
    'oracle': 'Oracle',
    'sap': 'SAP'
}

def normalize_company(name):
    if pd.isnull(name):
        return name
    name_lower = name.lower()
    for key, value in company_map.items():
        if key in name_lower:
            return value
    return name

# تطبيق التطبيع
df['company_name'] = df['company_name'].apply(normalize_company)
```

```python
import country_converter as coco


# U.S. state abbreviation to full name
us_state_full_names = {
    'AL': 'Alabama', 'AK': 'Alaska', 'AZ': 'Arizona', 'AR': 'Arkansas', 'CA': 'California',
    'CO': 'Colorado', 'CT': 'Connecticut', 'DE': 'Delaware', 'FL': 'Florida', 'GA': 'Georgia',
    'HI': 'Hawaii', 'ID': 'Idaho', 'IL': 'Illinois', 'IN': 'Indiana', 'IA': 'Iowa',
    'KS': 'Kansas', 'KY': 'Kentucky', 'LA': 'Louisiana', 'ME': 'Maine', 'MD': 'Maryland',
    'MA': 'Massachusetts', 'MI': 'Michigan', 'MN': 'Minnesota', 'MS': 'Mississippi',
    'MO': 'Missouri', 'MT': 'Montana', 'NE': 'Nebraska', 'NV': 'Nevada', 'NH': 'New Hampshire',
    'NJ': 'New Jersey', 'NM': 'New Mexico', 'NY': 'New York', 'NC': 'North Carolina',
    'ND': 'North Dakota', 'OH': 'Ohio', 'OK': 'Oklahoma', 'OR': 'Oregon', 'PA': 'Pennsylvania',
    'RI': 'Rhode Island', 'SC': 'South Carolina', 'SD': 'South Dakota', 'TN': 'Tennessee',
    'TX': 'Texas', 'UT': 'Utah', 'VT': 'Vermont', 'VA': 'Virginia', 'WA': 'Washington',
    'WV': 'West Virginia', 'WI': 'Wisconsin', 'WY': 'Wyoming', 'DC': 'District of Columbia'
}

# Country converter instance
cc = coco.CountryConverter()

# Clean and extract city
def extract_city(loc):
    if pd.isna(loc):
        return None
    parts = [p.strip() for p in str(loc).split(' ')]
    return parts[0] if parts else None

# Extract country
def extract_country(loc):
    if pd.isna(loc):
        return None

    parts = [p.strip() for p in str(loc).split(' ')]

    for part in reversed(parts):
        if part in us_state_full_names:
            return "United States"

    for part in reversed(parts):
        country = cc.convert(names=part, to='name_short')
        if country != 'not found':
            return country

    return None

# Apply
df['city'] = df['location'].apply(extract_city)
df['country'] = df['location'].apply(extract_country)
```

```python
silhouette_scores = {}

for model_name, model in models.items():
    try:
        # GaussianMixture بيحتاج fit_predict
        if model_name == 'GaussianMixture':
            labels = model.fit_predict(X_normalized)
        else:
            labels = model.fit(X_normalized).labels_

        # نتأكد إن فيه أكتر من Cluster
        if len(set(labels)) > 1:
            score = silhouette_score(X_normalized, labels)
            silhouette_scores[model_name] = score
            print(f"{model_name} Silhouette Score: {score:.4f}")
        else:
            print(f"{model_name}: Only one cluster found. Silhouette Score skipped.")

        print("="*50)
    except Exception as e:
        print(f"{model_name}: Error - {e}")
        print("="*50)
```

```python
# PCA
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_normalized)
```

```python
silhouette_scores = {}

for model_name, model in models.items():
    try:
        if model_name == 'GaussianMixture':
            labels = model.fit_predict(X_pca)
        else:
            labels = model.fit(X_pca).labels_

        if len(set(labels)) > 1:
            score = silhouette_score(X_pca, labels)
            silhouette_scores[model_name] = score
            print(f"{model_name} Silhouette Score: {score:.4f}")
        else:
            print(f"{model_name}: Only one cluster found.")

        print("="*50)
    except Exception as e:
        print(f"{model_name}: Error - {e}")
        print("="*50)
```

```python
from sklearn.model_selection import ParameterSampler
# --- Parameter grids for each model ---
param_grids = {
    'KMeans': {
        'n_clusters': [2, 3, 4, 5, 6, 7, 8],
        'init': ['k-means++', 'random'],
        'max_iter': [100, 200, 300, 500],
        'n_init': [10, 20, 30]
    },
    'AgglomerativeClustering': {
        'n_clusters': [2, 3, 4, 5, 6, 7, 8],
        'linkage': ['ward', 'complete', 'average', 'single']
    },
    'DBSCAN': {
        'eps': [0.3, 0.5, 0.7, 1.0, 1.5],
        'min_samples': [3, 5, 10]
    },
    'MeanShift': {
        'bandwidth': [None, 1, 2, 3]
    },
    'GaussianMixture': {
        'n_components': [2, 3, 4, 5, 6, 7, 8],
        'covariance_type': ['full', 'tied', 'diag', 'spherical']
    }
}
```

```python
best_results = {}

# --- Manual Random Search for each model ---
for model_name, grid in param_grids.items():
    print(f"\n=== {model_name} ===")
    param_list = list(ParameterSampler(grid, n_iter=10, random_state=42))
    best_score = -1
    best_params = None

    for params in param_list:
        try:
            if model_name == 'KMeans':
                model = KMeans(**params, random_state=42)
                labels = model.fit_predict(X_pca)
            elif model_name == 'AgglomerativeClustering':
                model = AgglomerativeClustering(**params)
                labels = model.fit(X_pca).labels_
            elif model_name == 'DBSCAN':
                model = DBSCAN(**params)
                labels = model.fit_predict(X_pca)
            elif model_name == 'MeanShift':
                model = MeanShift(**params)
                labels = model.fit_predict(X_pca)
            elif model_name == 'GaussianMixture':
                model = GaussianMixture(**params, random_state=42)
                labels = model.fit_predict(X_pca)
            else:
                continue

            # Skip if only one cluster is found
            if len(set(labels)) > 1:
                score = silhouette_score(X_pca, labels)
                if score > best_score:
                    best_score = score
                    best_params = params

        except Exception as e:
            print(f"Error with params {params}: {e}")

    best_results[model_name] = {'best_params': best_params, 'best_score': best_score}
    print(f"Best Params: {best_params}")
    print(f"Best Silhouette Score: {best_score:.4f}")

print("\n=== Final Best Results ===")
for model, result in best_results.items():
    print(f"{model}: Score={result['best_score']:.4f}, Params={result['best_params']}")
```

```python
#best model get heigher score in best_result
best_model = max(best_results, key=lambda k: best_results[k]['best_score'])
print(f"The best model is: {best_model}")
```

Common models used for clustering include:

*   **K-Means Clustering:** An iterative algorithm that aims to partition the data into k clusters. It works by minimizing the within-cluster sum of squares.
*   **Hierarchical Clustering:** This method builds a hierarchy of clusters. It can be either agglomerative (bottom-up) or divisive (top-down).
*   **DBSCAN (Density-Based Spatial Clustering of Applications with Noise):** This algorithm groups together points that are closely packed together, identifying outliers as points in low-density regions.
*   **Mean Shift:** A non-parametric algorithm that finds modes of density in the data.
*   **Gaussian Mixture Models (GMM):** A probabilistic model that assumes the data is generated from a mixture of several Gaussian distributions.

The choice of clustering model depends on the nature of your data and the desired outcome.
